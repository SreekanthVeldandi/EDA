```python
## Compute the X and Y coordinates for points on sin curve

x=np.arange(0,4)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.arange(0,6*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
# subplot
#Compute the X and Y Coordintes for points on sin and cosine curves

x=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

# set up a subplot that has a height=2 and width=1,
plt.subplot(2, 1, 1)

#Make the first plot as active
plt.plot(x,y_sin,'r--')
plt.title('sine')

plt.subplot(2,1,2)
#Set the second sub plot as active
plt.plot(x,y_cos,'bo')
plt.title('cosine')

plt.show()
```

# Bar plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[2,8,10]
y=[11,16,19]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar graph')
```

# Boxplot

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

#Rectangular Boxplot
plt.boxplot(data,vert='True',patch_artist='False')
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert='False',patch_artist='True')
```
